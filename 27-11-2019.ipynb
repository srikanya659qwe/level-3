```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,x*x,'ro')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('line')
plt.subplot(2,1,2)
plt.plot(x,x*x,'ro')
plt.title('y=x * x')
plt.tight_layout()
```
